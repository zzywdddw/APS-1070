## 2.NumPy
NumPy is the fundamental package for scientific computing in Python. It is a Python library that provides a multidimensional array object, various derived objects (such as masked arrays and matrices), and an assortment of routines for fast operations on arrays, including mathematical, logical, shape manipulation, sorting, selecting, I/O, discrete Fourier transforms, basic linear algebra, basic statistical operations, random simulation and much more.

###Why is NumPy Fast?

Vectorization describes the absence of any explicit looping, indexing, etc., in the code - these things are taking place, of course, just “behind the scenes” in optimized, pre-compiled C code. Vectorized code has many advantages, among which are:

* vectorized code is more concise and easier to read

* fewer lines of code generally means fewer bugs
* the code more closely resembles standard mathematical notation (making it easier, typically, to correctly code mathematical constructs)
* vectorization results in more “Pythonic” code. Without vectorization, our code would be littered with inefficient and difficult to read for loops.

### Significance of vectorized coding:

Writing loops to handle different scenarios in our algorithms might seem the most straightforward solution; but it is not the efficient one. In the next few cells, we will try to do an element-wise vector multiplication on A and B vectors. Each vector has 10 million elements. First, we will multiply each element of A with the corresponding element of B and store the result in C using a for loop. Then we try to implement it by vectorized operation only as `A * B`. Let's see what happens!

```
A = np.arange(0,10000000,1)
```

```
A
```

```
B = np.random.random(10000000)
C1 = np.zeros(10000000)
C2 = np.zeros(10000000)
```

```
import time

start=time.time()

for i in range (0,10000000):
    C1[i]=A[i]*B[i]

elapsed_time1 = time.time()-start

print("For loop Elapsed time: ", elapsed_time1)
```

```
start=time.time()

C2 = A * B

elapsed_time2 = time.time()-start

print("Vectorized Elapsed time: ", elapsed_time2)
```

```
Error=np.sum(abs(C2-C1))
print("Error=", Error)
print("Speedup=", elapsed_time1/elapsed_time2)
```

As expected the vectoried computation was faster, but how much? Like $2\times$? or wait... $200\times$!! We should definitely try to avoid loops as much as possible!

### Subplots

```
import matplotlib.pyplot as plt
import numpy as np

# Some example data to display
x = np.linspace(0, 2 * np.pi, 400)
y = np.cos(x ** 2)
```

```
fig, axs = plt.subplots(2)
fig.suptitle('Vertically stacked subplots')
axs[0].plot(x, y , 'r')
axs[1].plot(x, -y)
```

```
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, y)
axs[0, 0].set_title('Axis [0,0]')
axs[0, 1].plot(x, y**2, 'tab:orange')
axs[0, 1].set_title('Axis [0,1]')
axs[1, 0].plot(x, -y**3, 'tab:green')
axs[1, 0].set_title('Axis [1,0]')
axs[1, 1].plot(x, -y**4, 'tab:red')
axs[1, 1].set_title('Axis [1,1]')

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```

```
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)

fig.set_size_inches(10, 5) # Set figure size

fig.suptitle('Sharing x per column, y per row')
ax1.plot(x, y)
ax2.plot(x, y**2, 'tab:orange')
ax3.plot(x, -y**3, 'tab:green')
ax4.plot(x, -y**4, 'tab:red')

for ax in fig.get_axes():
    ax.label_outer()
```
